```python
import gymnasium as gym
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import trading_environment
```

```python
def generate_random_df(n: int = 1000):
    df = np.empty(n)
    # Starting value
    df[0] = 100

    for i in range(1, n):
        df[i] = df[i - 1] * (1 + np.random.random() * 0.2 - 0.1)

    return pd.DataFrame({'Close': df})


df = generate_random_df()
```

```python
env = gym.make("trading/csv", df=df)
```

```python
agents = []

for _ in range(1000):
    env.reset()
    net_worth_history = []

    for i in range(100):
        *_, info = env.step(env.action_space.sample())
        net_worth_history.append(info['net_worth'])
    agents.append(net_worth_history)
```
